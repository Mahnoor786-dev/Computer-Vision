```python
import numpy as np
from PIL import Image
import matplotlib.pylab as plt
img=Image.open("cat.jpg")
img_arr=img.load()
width=img.size[0]
height=img.size[1]
for col in range(width):
  for row in range(height):
    pixel = img_arr[col,row]
    avg=int(0.3*pixel[0] + 0.59*pixel[1] + 0.11*pixel[2])
    img_arr[col,row]=(avg,avg,avg)

display(img)
```
